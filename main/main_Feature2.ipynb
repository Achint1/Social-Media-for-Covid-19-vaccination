```python
df_dictionary = pd.read_csv('gdrive/MyDrive/researchProjectIME/data_dictionary.csv')
df_dictionary.head()
```

```python
keywords = []
for word in df_dictionary.making:
  keywords.append(word)
```

```python
keywords
```

```python
df.head()
```

```python
df['frequency']=0
for word in keywords:
  df['frequency'] += df['tweet'].map(lambda x: x.count(word))
```

```python
df.head()
size=df.shape[0]
```

```python
sum=[]
for i in range(31):
  s=0
  for j in range(size):
    if(df['dates'][j]==i):
      s=s+df['frequency'][j]
  sum.append(s)
```
